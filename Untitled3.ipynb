```python
import matplotlib.pyplot as plt
plt.style.use('seaborn-poster')
%matplotlib inline

import numpy as np


import pandas as pd
```

# pT classification preformance

```python
path = '/mnt/t3nfs01/data01/shome/jandrejk/higgs_model_dep/MoriondAnalysis/differential_recoMatrices/'
names = ['3clfs_std', '3clfs_AngGamJetGamGamJetJetClass_AngGamJetGamGamJetJetrecoPt_AngGamJetGamGamJetJetrecoNjets',
        '3clfs_AngGamJetGamGamJetJetClass_DistRrecoPt_DistR_recoNjets']


for process in ['ggF','VBF','VH','ttH','SM'] :
    #process = 'ggF'
    N_true = []
    N_err = []
    N_pred = []

    for name in names :
        mat = np.load(path+'/'+name+'/recoPt/'+'N_reco_true_'+process+'.npy')
        N_true.append(np.sum(mat,axis=1))

        mat = np.load(path+'/'+name+'/recoPt/'+'N_err_reco_true_'+process+'.npy')
        mat = np.nan_to_num(mat)
        mat2 = np.square(mat)

        N_err.append(np.sqrt(np.sum(mat2,axis=1)))

        mat = np.load(path+'/'+name+'/recoPt/'+'N_reco_pred_'+process+'.npy')
        N_pred.append(np.sum(mat,axis=1))


    std = (N_pred[0]-N_true[0])/N_err[0]
    angular = (N_pred[1]-N_true[1])/N_err[1] 
    distR = (N_pred[2]-N_true[2])/N_err[2]


    plt.hist(std,bins=np.linspace(-2,2,10),alpha=0.5,label='std')
    plt.hist(angular,bins=np.linspace(-2,2,10),alpha=0.5,label='angular')
    plt.hist(distR,bins=np.linspace(-2,2,10),alpha=0.5,label=r'$\Delta R$')

    print '-----------------------'
    print 'process: ', process
    print 'std'
    print np.mean(std)
    print np.var(std,ddof=-1)
    print 'angular'
    print np.mean(angular)
    print np.var(angular,ddof=-1)
    print 'distR'
    print np.mean(distR)
    print np.var(distR,ddof=-1)
    print '-----------------------'
    plt.legend()
    plt.show()
```

```python
path = '/mnt/t3nfs01/data01/shome/jandrejk/higgs_model_dep/MoriondAnalysis/differential_recoMatrices/'
names = ['3clfs_std', '3clfs_AngGamJetGamGamJetJetClass_AngGamJetGamGamJetJetrecoPt_AngGamJetGamGamJetJetrecoNjets',
        '3clfs_AngGamJetGamGamJetJetClass_DistRrecoPt_DistR_recoNjets',
         '2clfs_AngGamJetGamGamJetJetClass_AntiKTrecoNjets',
        'pt20GeVcut']


for process in ['ggF','VBF','VH','ttH','SM'] :
    #process = 'ggF'
    N_true = []
    N_err = []
    N_pred = []

    for name in names :
        mat = np.load(path+'/'+name+'/recoNjets/'+'N_reco_true_'+process+'.npy')
        N_true.append(np.sum(mat,axis=1))

        mat = np.load(path+'/'+name+'/recoNjets/'+'N_err_reco_true_'+process+'.npy')
        mat = np.nan_to_num(mat)
        mat2 = np.square(mat)

        N_err.append(np.sqrt(np.sum(mat2,axis=1)))

        mat = np.load(path+'/'+name+'/recoNjets/'+'N_reco_pred_'+process+'.npy')
        N_pred.append(np.sum(mat,axis=1))


    std = (N_pred[0]-N_true[0])/N_err[0]
    angular = (N_pred[1]-N_true[1])/N_err[1] 
    distR = (N_pred[2]-N_true[2])/N_err[2]
    anti_Kt = (N_pred[3]-N_true[3])/N_err[3]
    pt20cut = (N_pred[4]-N_true[4])/N_err[4]

    plt.hist(std,bins=np.linspace(-2,2,10),alpha=0.5,label='std')
    plt.hist(angular,bins=np.linspace(-2,2,10),alpha=0.5,label='angular')
    plt.hist(distR,bins=np.linspace(-2,2,10),alpha=0.5,label=r'$\Delta R$')
    plt.hist(anti_Kt,bins=np.linspace(-2,2,10),alpha=0.5,label=r'anti-kT')
    plt.hist(pt20cut,bins=np.linspace(-2,2,10),alpha=0.5,label=r'jetpT20')

    print '-----------------------'
    print 'process: ', process
    print 'std'
    print np.mean(std)
    print np.var(std,ddof=-1)
    print 'angular'
    print np.mean(angular)
    print np.var(angular,ddof=-1)
    print 'distR'
    print np.mean(distR)
    print np.var(distR,ddof=-1)
    print 'anti_Kt'
    print np.mean(anti_Kt)
    print np.var(anti_Kt,ddof=-1)
    print 'pt20cut'
    print np.mean(pt20cut)
    print np.var(pt20cut,ddof=-1)
    print '-----------------------'
    plt.legend()
    plt.show()
```
